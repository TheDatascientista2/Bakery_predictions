```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load data
df = pd.read_csv('merged_data_clean.csv', parse_dates=['Datum'])
```

```python
# Replace NaN with 0 for clarity (if not already done)
df['KielerWoche'] = df['KielerWoche'].fillna(0)

# Group by 'KielerWoche' and calculate mean sales
df_kielerwoche_sales = df.groupby('KielerWoche')['Umsatz'].mean().reset_index()

# Plot
plt.figure(figsize=(8, 5))
plt.bar(df_kielerwoche_sales['KielerWoche'], df_kielerwoche_sales['Umsatz'], color=['#b09fb3' if x == 0 else '#a2b39f' for x in df_kielerwoche_sales['KielerWoche']])
plt.title('Average Sale: KiWo vs. not KiWo')
#plt.xlabel('Kieler Woche (0 = Nein, 1 = Ja)')
plt.ylabel('Average Sales [€]')
plt.xticks([0, 1], ['No KiWo', 'During KiWo'])
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
